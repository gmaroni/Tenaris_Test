## Informazione mutua

```python
from sklearn.feature_selection import mutual_info_classif, mutual_info_regression
```

Mutual information measures how much information the presence/absence of a feature contributes to making the correct prediction on Y.

As extracted from [wikipedia](https://en.wikipedia.org/wiki/Mutual_information):

Mutual information measures the information that X and Y share: It measures how much knowing one of these variables reduces uncertainty about the other. For example, if X and Y are independent, then knowing X does not give any information about Y and vice versa, so their mutual information is zero. At the other extreme, if X is a deterministic function of Y and Y is a deterministic function of X then all information conveyed by X is shared with Y: knowing X determines the value of Y and vice versa. As a result, in this case the mutual information is the same as the uncertainty contained in Y (or X) alone, namely the entropy of Y (or X). Moreover, this mutual information is the same as the entropy of X and as the entropy of Y. (A very special case of this is when X and Y are the same random variable.)


```python
y = df.pe.copy()
X = df.drop('pe',1).copy()
```

```python
mi = mutual_info_regression(X, y)
mi = pd.Series(mi)
mi.index = X.columns
```

```python
mi.sort_values(ascending=False).plot.bar(figsize=(12,6), fontsize=16)
```

## Target binning

Adesso implemento un metodo utilizzato nella competizione KDD 2019 utilizzata dai data scientist dell'università di melbourne vincitori di tale competizione 

```python
def KDD_feat_imp(df, var, target, n_bins=100):
    
    labels = ['Q' + str(i + 1) for i in range(0, n_bins)]

    df[f'{var}_binned'], intervals = pd.qcut(
        df[var],
        n_bins,
        labels=labels,
        retbins=True,
        precision=3,
        duplicates='drop')
    
    risk_dict = df.groupby([f'{var}_binned'])[target].mean().to_dict()

    df[f'{var}_binned'] = df[f'{var}_binned'].map(risk_dict)
    return df
```

```python
from sklearn.metrics import r2_score
r2 = pd.Series(index=df.drop('pe',axis=1).columns)

df2 = KDD_feat_imp(df, 'at', 'pe', n_bins=100)

r2['at']=r2_score(df.pe, df2['at_binned'])
```

```python
df2 = KDD_feat_imp(df, 'v', 'pe', n_bins=100)

r2['v']=r2_score(df.pe, df2['v_binned'])
```

```python
df2 = KDD_feat_imp(df, 'ap', 'pe', n_bins=100)

r2['ap']=r2_score(df.pe, df2['ap_binned'])
```

```python
df2 = KDD_feat_imp(df, 'rh', 'pe', n_bins=100)

r2['rh']=r2_score(df.pe, df2['rh_binned'])
```

```python
r2.sort_values(ascending=False).plot.bar(figsize=(12,6), fontsize=16)
```

## Lasso regression

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso

def cv_optimize_lasso(X, y, list_of_lambdas, n_folds=4):
    clf = Lasso(normalize=True)
    parameters = {"alpha": list_of_lambdas}
    gs = GridSearchCV(clf, param_grid=parameters, cv=n_folds, scoring="neg_mean_squared_error")
    gs.fit(X, y)
    return gs
```

```python
lol = [1e-8,1e-6, 1e-5, 5e-5, 1e-4, 5e-4, 1e-3, 1e-2, 1e-1, 1.0, 10.0]

fitmodel_lasso = cv_optimize_lasso(X, y, lol, n_folds=4)
```

```python
lambdawechoose_lasso = fitmodel_lasso.best_params_['alpha']

est_lasso = Lasso(alpha=lambdawechoose_lasso, normalize=True).fit(X,y)
est_lasso
```

```python
def nonzero_lasso(est, lcols):
    featuremask=(est.coef_ !=0.0)
    return pd.DataFrame(dict(feature=lcols, coef=est.coef_, 
        abscoef=np.abs(est.coef_)))[featuremask].sort_values('abscoef',ascending=False)
```

```python
d=3

lasso_importances=nonzero_lasso(est_lasso, X.columns)
lasso_importances.set_index("feature", inplace=True)
lasso_importances
```

```python
lasso_importances.abscoef.sort_values(ascending=False).plot.bar(figsize=(12,6), fontsize=16)
```

#### Tutti i metodi confermano il medesimo risultato e confermano le ipotesi fatte durante la fase di analisi esplorativa ovvero il fatto che la variabile at è la variabile più importante per predire la target pe, e a seguire vi è la variabile v. 

#### Il metodo basato sulla regressione lineare con penalizzazione della norma 1 dei coefficienti (Lasso) sembra allocare una importanza relativa molto più alta alla variabile at rispetto che alle altre. Ciò è in realtà u risultato atteso in quanto questo metodo promuove la sparsità della soluzione e di conseguenza tende a portare a zero tutte le features a parte quelle più rilevanti all'aumentare della forza di regolarizzazione 

# Feature importance (non lineare)

Tutti i metodi di cui sopra sono utili nel caso in cui si assume che vi sia una dipendenza lineare tra target e features. Nei casi reali però si osservano spesso relazioni non lineari tra features e target che non sono in gradi di essere apprezzate dai metodi usati sopra, sopratutto nel caso in cui la target dipenda da una interazione non lineare tra più features di ingresso.

Per scoprire eventuali relazioni non lineari tra features e target vi sono diversi modi. 

Il modo più immediato è quello di trasformare con trasformazioni non lineari le features e poi utilizzare i metodi usati sopra.

Un altro metodo è quello di utilizzare dei modelli di ml non lineari come per esempio metodi basati su alberi e utilzzare le tecniche di feature importance basate sul potere predittivo delle singole features rispetto alla target

## Metodo di ricerca esaustivo basato su random forest

Un ultimo modo per calcolare l'importanza delle features è selezionare un modello come per esempio un Random Forest, una metrica di valutazione come per esempio la metrica r2 e calcolare l'importanza predittiva delle features allenando il RF con una features singola alla volta e osservare con quale features si ottiene il massimo del potere predittivo della target.

Il metodo sottostante utilizza un RF per effettuare una ricerca esaustiva della miglior combinazione di numero di features per predire la target, utilizzo questo metodo per avere sia una informazione della feature importance dal punto di vista di un RF (ma mi bastava utilizzare la funzione feature_importance) sia per verificare la miglior combinazione di features dal punto di vista del potere prdittivo, nel caso in cui la miglior combinazione escluda una o più features ciò ignifica che le features escluse non hanno potere predittivo e anzi rappresentano rumore (in realtà da tutte le considerazioni fatte in precedenza so già che ciò non si verificherà in quanto tutte e 4 le features sembrano avere una correlazione con la target)

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
from sklearn.ensemble import RandomForestRegressor


efs1 = EFS(RandomForestRegressor(n_jobs=-1, n_estimators=200), 
           min_features=1,
           max_features=1,
           scoring='r2',
           print_progress=True,
           cv=2)

efs1 = efs1.fit(X, y)
```

```python
efs1.best_feature_names_, efs1.best_score_
```

In questo caso, a differenza dei casi precedenti la features più predittiva sembra essere la variabile v.

Occorre però prestare attenzione a questo risultato, in quanto questa metodologia è basata sul modello e sulla metrica scelti, cambiando modello e cambiando metrica si possono ottenere risultati differenti. Inoltre ciò può essere dovuto al fatto che le due variiabili at e v posseggono una forte correlazione e di conseguenza il modello risulta 'ingannato' rispetto a quale sia la feature effettivamente più importante

```python
efs1 = EFS(RandomForestRegressor(n_jobs=4), 
           min_features=1,
           max_features=2, 
           scoring='r2',
           print_progress=True,
           cv=2)

efs1 = efs1.fit(X, y)
```

```python
efs1.best_feature_names_, efs1.best_score_
```

```python
efs1 = EFS(RandomForestRegressor(n_jobs=4), 
           min_features=1,
           max_features=3, 
           scoring='r2',
           print_progress=True,
           cv=2)

efs1 = efs1.fit(X, y)
```

```python
efs1.best_feature_names_, efs1.best_score_
```

```python
efs1 = EFS(RandomForestRegressor(n_jobs=4), 
           min_features=1,
           max_features=4, 
           scoring='r2',
           print_progress=True,
           cv=2)

efs1 = efs1.fit(X, y)
```

```python
efs1.best_feature_names_, efs1.best_score_
```

Le features ap ed rh sono considerate meno importanti dal RF, però tuttavia dimostrano potere predittivo dal momento che utilizzare tutte e 4 le features da il risultato migliore in termini di metrica r2 

## Considerazioni finali

#### L'importanza di una feature e il suo potere predittivo dipendono da tutta una serie di assunzioni, in primo luogo la relazione ce si sta cercando nei dati (lineare/nonlineare), il modello che si vuole utilizzare per fittare i dati e la metrica che si vuole ottimizzare.

#### Da questo notebook risulta evidente che la feature at e a seguire la feature v sono le features più correlate linearmente con la target, sono di conseguenza quelle più informative se si vuole utilizare un modello lineare.

#### Anche utilizzando un modello nonlineare la coppia di features at e v risulta quella più informativa
