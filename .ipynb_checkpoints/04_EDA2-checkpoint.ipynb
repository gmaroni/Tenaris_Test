# Analisi esplorativa

Dal momento che dall'analisi precedente abbiamo verificato che i segnali in gioco sono iid in questo notebook verrà effettuata una analisi esplorativa più classica rispetto a quella del notebook precedente 

```python
df = pd.read_csv('data/data02.csv', low_memory=False, index_col=0)
df
```

## Correlazione tra gli input e la target

```python
ax = sns.scatterplot(x="at", y="pe",hue="pe", data=df)
```

Si osserva una forte correlazione lineare negativa tra la variabile at e la variabile target

```python
ax = sns.scatterplot(x="v", y="pe",hue="pe", data=df)
```

Si osserva una relazione non lineare tra v e pe, in particolare una relazione a iperbole, infatti trasformando la variabile v in modo opportuno è possibile ottenere una relazione lineare tra v e pe

```python
ax = sns.scatterplot(x=1.0/df.v, y="pe",hue="pe", data=df)
```

```python
ax = sns.scatterplot(x=df.ap, y="pe",hue="pe", data=df)
```

Non sembra esserci una forte correlazione lineare tra ap e pe

```python
ax = sns.scatterplot(x=df.rh, y="pe",hue="pe", data=df)
```

Non sembra esserci una forte correlazione lineare tra rh e pe

Per quantificare la correlazione tra tutte le variabili in gioco è possibile calcolare la matrice di correlazione

```python
plt.figure(figsize=(7,6))
ax=sns.heatmap(df.corr(), annot=True, fmt=".2f")
```

Trasformando la variabile v nella sua inversa si osserva un leggero aumento della sua correlazione lineare con pe

```python
df2 = df.copy()
```

```python
df2.v = 1.0/df2.v
```

```python
plt.figure(figsize=(7,6))
ax=sns.heatmap(df2.corr(), annot=True, fmt=".2f")
```

Si osserva che la correlazione tra at e pe e tra v e pe è talmente forte che si presuppone che pe sia stata generata come combinazione lineare di at e v

Osserviamo, tramite un pairplot tutte le combinazioni di correlazioni tra tutte le variabili del dataset.

```python
sns.pairplot(df);
```

#### La mia supposizione osservando le distribuzioni empiriche in gioco è che le variabili rh e ap siano state prodotte per prime attraverso un meccanismo di generazione di variabili casuali. Ap è stata generata semplicemente come una gaussina con media circa 1013 e deviazione standard di circa 6, mentre la varibile rh come una distribuzione con skew negativa. A partire da queste sono state poi generate le variabili at e v (oppure v è stata generata a posteriori da at) attraverso qualche combinazione oppportuna di ap ed rh che ne spiegherebbe la correlazione.

#### Detto questo utilizziamo una serie di metodi per calcolare la feature importance delle variabili di ingresso del dataset e verifichiamo che le ipotesi siano plausibili
